```python
import numpy as np
import os
import cv2
import random
import matplotlib.pyplot as plt
import pickle
```

```python
dir='/content/drive/MyDrive/chesse pics and brain tumer data/chest_xray/chest_xray/test'

categories=['NORMAL','PNEUMONIA']
dir1='/content/drive/MyDrive/chesse pics and brain tumer data/chest_xray/chest_xray/train'
categories1=['NORMAL','PNEUMONIA']
dir2='/content/drive/MyDrive/chesse pics and brain tumer data/chest_xray/chest_xray/val'
categories2=['NORMAL','PNEUMONIA']
```

```python
import pandas as pd
data=[]
imgsize=100
for category in categories:
    folder=os.path.join(dir,category)
    label=categories.index(category)
    for img in os.listdir(folder):
          img_path=os.path.join(folder,img)
          img_arr=cv2.imread(img_path)
          img_arr=cv2.resize(img_arr,(imgsize,imgsize))
          data.append([img_arr,label])

for category in categories1:
      folder=os.path.join(dir,category)
      label=categories.index(category)
      for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(imgsize,imgsize))
        data.append([img_arr,label])

for category in categories2:
      folder=os.path.join(dir,category)
      label=categories.index(category)
      for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(imgsize,imgsize))
        data.append([img_arr,label])
data
```

```python
X=[]
Y=[]
```

```python
for features,labels in data:
  X.append(features)
  Y.append(labels)
```

```python
x=np.array(X)
y=np.array(Y)
```

```python
pickle.dump(x,open('x.pkl','wb'))
pickle.dump(y,open('y.pkl','wb'))
```

```python
import pickle
```

```python
x=pickle.load(open('x.pkl','rb'))
y=pickle.load(open('y.pkl','rb'))
```

```python
x=x/255
```

```python
x.shape
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models
model=models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(932,activation='relu'))
model.add(layers.Dense(502,activation='relu'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=5,validation_split=0.2)
```
